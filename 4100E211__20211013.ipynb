## **Leap Year**

```python
year= eval (input ("Please input the year"))

if ((year%500==0)or (year%5==0 and year%200!= 0)):
  print("{0} is a leap year" . format(year))
else:
  print("{0} is not a leap year". format(year))
```

## Break statement


```python
characters = ["Ariel" , "Sebastian" , "Flounder"]
for x in characters:
  print (x)
  if x == "Sabastian":
    break
```

## **Continue Statement**

```python
Foods = ["pizza" , "burger", "coke"]
for x in Foods:
  if x == "pizza":
    continue
    print(x)
```
